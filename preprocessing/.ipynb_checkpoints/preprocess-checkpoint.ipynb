```python
import pandas as pd
import numpy as np
```

```python
# ten nations we are interested in.
nations = ['United States', 'China', 'Russian Federation', 'Germany',
           'United Kingdom', 'France', 'Italy', 'Iran, Islamic Rep.', 
           'Israel', 'Saudi Arabia']
# time span
years = ['2012', '2013', '2014', '2015', '2016']
```

```python
# read data of military spending (million USD)
rawdata_ms = pd.read_csv('military.csv',skiprows=3,index_col=0)
ms = rawdata_ms.loc[nations,years] #select data we need
ms = (ms[years]/10**6).astype('int64') #get rid of the scientific notation, million USD
ms
```

```python
# read data of gdp (million USD)
rawdata_gdp = pd.read_csv('gdp.csv',skiprows=3,index_col=0)
gdp = rawdata_gdp.loc[nations,years]
gdp = (gdp[years]/10**6).astype('int64')
gdp
```

```python
# read data of education (million USD)
rawdata_health = pd.read_csv('health.csv',skiprows=3,index_col=0)
health = rawdata_health.loc[nations,years]
health = (health/100*gdp).astype('int64')
health
```

```python
# read data of education (million USD)
rawdata_edu = pd.read_csv('education.csv',skiprows=3,index_col=0)
edu = rawdata_edu.loc[nations,years]
edu = edu.fillna(0)
edu = (edu/100*gdp).astype('int64')
edu = edu.replace(0, np.nan)
edu
```

```python
# read data of population (million)
rawdata_pop = pd.read_csv('population.csv',skiprows=3,index_col=0)
pop = rawdata_pop.loc[nations,years]
pop = (pop[years]/10**6).astype('float64')
pop
```

```python
def changeIndexName(df, newName):
    df.index = newName
    return df
```

```python
new_nations = ['US', 'China', 'Russia', 'Germany',
           'UK', 'France', 'Italy', 'Iran', 
           'Israel', 'Saudi Arabia']
```

```python
changeIndexName(ms, new_nations)
changeIndexName(gdp, new_nations)
changeIndexName(health, new_nations)
changeIndexName(edu, new_nations)
changeIndexName(pop, new_nations)
```

```python
ms.to_csv('cleaneddata/ms.csv')
gdp.to_csv('cleaneddata/gdp.csv')
health.to_csv('cleaneddata/health.csv')
edu.to_csv('cleaneddata/edu.csv')
pop.to_csv('cleaneddata/pop.csv')
```

```python
pcms = ms/pop
pcgdp = gdp/pop
pchealth = health/pop
pcedu = edu/pop
```

```python
pcms.to_csv('cleaneddata/pcms.csv')
pcgdp.to_csv('cleaneddata/pcgdp.csv')
pchealth.to_csv('cleaneddata/pchealth.csv')
pcedu.to_csv('cleaneddata/pcedu.csv')
```

```python
health_past = health.iloc[:,:-1]
health_now = health.iloc[:,1:]
growth_health = np.subtract(health_now, health_past)
growth_health_pct = np.divide(growth_health, health_past)
```

```python
edu_past = edu.iloc[:,:-1]
edu_now = edu.iloc[:,1:]
growth_edu = np.subtract(edu_now, edu_past)
growth_edu_pct = np.divide(growth_edu, edu_past)
```

```python
growth_edu.to_csv('cleaneddata/growth_edu.csv')
growth_edu_pct.to_csv('cleaneddata/growth_edu_pct.csv')
growth_health.to_csv('cleaneddata/growth_health.csv')
growth_health_pct.to_csv('cleaneddata/growth_health_pct.csv')
```
